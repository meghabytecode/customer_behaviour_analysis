```python
import pandas as pd
import mysql.connector
```

```python
#Loading csv file
df = pd.read_csv(r"E:\learnings\New folder\customer_shopping_behavior.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#check for null values
df.isnull().sum()
```

```python
#checking for complete data conting null values
df[df['Review Rating'].isna()]
```

```python
#filling null values by median of individual category
df['Review Rating'] = df.groupby('Category')['Review Rating'] .transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#making column name to snake case
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
df.columns
```

```python
# age group into 4 catagory
df['age_group'] = pd.qcut(df['age'], q=4, labels = ['Young adults', 'Adults', 'Middle aged', 'Senior'])
```

```python
# extra
df['age_group'].value_counts()
```

```python
# extra
_, bins = pd.qcut(df['age'], 4, retbins=True)
bins
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
freq_map = {
    'Weekly': 7,
    'Fortnightly': 14,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Annually': 365
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(freq_map)
```
